```python
def power_fuc(base: int, power: int) -> int:
    if not isinstance(base, int) or not isinstance(power, int):
        raise TypeError("Input must be digit.")

    result = 1
    for _ in range(power):
        result *= base
    return result


if __name__ == "__main__":
    try:
        base = input("Enter the base value: ")
        power = input("Enter the power: ")

        if base == "" or power == "":
            print("Empty")
        else:
            print(power_fuc(int(base), int(power)))

    except ValueError:
        print("Please enter valid input.")
    except TypeError as e:
        print(e)
```
